```python
data['Date']= pd.to_datetime(data['Date'])
```

```python
data["Date"] = data["Date"].dt.date
```

```python
import datetime
snapshot_date = max(data.Date) + datetime.timedelta(days=1)
```

```python
customers = data.groupby(['uuid_Cat']).agg({
    'Date': lambda x: (snapshot_date - x.max()).days,
    'uuid_Cat': 'count',
    'price': 'sum'})
```

```python
customers.rename(columns = {'Date': 'Recency',
                            'uuid_Cat': 'Frequency',
                            'Price': 'MonetaryValue'}, inplace=True)
```

```python
customers
```

```python
customers_fix = pd.DataFrame()
customers_fix["Recency"] = stats.boxcox(customers['Recency'])[0]
customers_fix["Frequency"] = stats.boxcox(customers['Frequency'])[0]
customers_fix["MonetaryValue"] = pd.Series(np.cbrt(customers['price'])).values
customers_fix.tail()
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(customers_fix)
customers_normalized = scaler.transform(customers_fix)
```

```python
print(customers_normalized.mean(axis = 0).round(2))
print(customers_normalized.std(axis = 0).round(2))
```

```python
scaler
```

```python
customers_normalized 
```

```python
sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(customers_normalized)
    sse[k] = kmeans.inertia_ # SSE to closest cluster centroid
plt.title('The Elbow Method')
plt.xlabel('k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```python
model = KMeans(n_clusters=3, random_state=42)
model.fit(customers_normalized)
model.labels_.shape
```

```python
customers["Cluster"] = model.labels_
customers.groupby('Cluster').agg({
    'Recency':'mean',
    'Frequency':'mean',
    'price':['mean', 'count']}).round(2)
```

```python
# Create the dataframe
df_normalized = pd.DataFrame(customers_normalized, columns=['Recency', 'Frequency', 'price'])
df_normalized['ID'] = customers.index
df_normalized['Cluster'] = model.labels_
# Melt The Data
df_nor_melt = pd.melt(df_normalized.reset_index(),
                      id_vars=['ID', 'Cluster'],
                      value_vars=['Recency','Frequency','price'],
                      var_name='Attribute',
                      value_name='Value')
df_nor_melt.head()
# Visualize it
sns.lineplot('Attribute', 'Value', hue='Cluster', data=df_nor_melt)
```
