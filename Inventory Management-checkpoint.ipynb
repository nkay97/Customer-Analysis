```python
count_product= new_data['category'].value_counts()
count_product.to_frame().head(60)
```

```python
temp = pd.DataFrame({"Categories": new_data['category']})
count_prod = temp.value_counts().head(10)
#print(count_prod)
labels = [X[0] for X in count_prod.keys()]


bar , ax = plt.subplots(figsize=(25,9))
ax = plt.pie(x = count_prod,autopct = "%.1f%%" ,labels = labels)
plt.title("Category Preferred",fontsize = 17)
plt.xticks(rotation = 45)
plt.show()
```

```python
count_product= new_data['mainproduct'].value_counts()
count_product.to_frame().head(60)
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x = data['mainproduct'].value_counts().head(25).values,
            y = data['mainproduct'].value_counts().head(25).index,
           data = data)
plt.title("Products attracted to Customers")
plt.xlabel("Product count")
plt.ylabel("Products") 
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.countplot(new_data['mainproduct'] [new_data['userevent'] == 'buy'], data=data)
plt.title("Products Purchased",fontsize = 17)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(20, 9))
sns.countplot(data['mainproduct'] [data['userevent'] == 'addToCart'], data=data)
plt.title("Products added to cart",fontsize = 17)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(20, 9))
sns.countplot(data['mainproduct'] [data['userevent'] == 'pageView'], data=data)
plt.title("Products viewed",fontsize = 17)
plt.xticks(rotation = 90)
plt.show()
```
